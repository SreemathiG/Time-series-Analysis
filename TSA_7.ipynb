```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
np.random.seed(42)
dates = pd.date_range(start="2020-01-01", periods=100, freq="D")
```

```python
trend = np.linspace(10, 100, 100)  # Linear trend
seasonality = 10 * np.sin(np.linspace(0, 3 * np.pi, 100))
noise = np.random.normal(0, 5, 100)  # Random noise
data = trend + seasonality + noise
```

```python
df = pd.DataFrame(data, index=dates, columns=["Value"])
```

```python
result = seasonal_decompose(df["Value"], model='additive', period=30)
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(411)
plt.plot(df.index, df["Value"], label="Original Data")
plt.title("Original Data")
plt.legend(loc='best')
```

```python
plt.subplot(412)
plt.plot(df.index, result.trend, label="Trend", color='red')
plt.title("Trend Component")
plt.legend(loc='best')
```

```python
plt.subplot(413)
plt.plot(df.index, result.seasonal, label="Seasonality", color='green')
plt.title("Seasonality Component")
plt.legend(loc='best')
```

```python
plt.subplot(414)
plt.plot(df.index, result.resid, label="Residuals", color='orange')
plt.title("Residual Component")
plt.legend(loc='best')
```

```python
plt.tight_layout()
plt.show()
```
